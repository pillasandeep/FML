```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
import nltk
nltk.download('punkt')
sentence="summering of removing importing"
word=PorterStemmer()
def stem_word(sentence):
  token_word=word_tokenize(sentence)
  x=[]
  for i in token_word:
    x.append(word.stem(i))
    x.append(" ")
  return " ".join(x)
y=stem_word(sentence)
print(y)
```

```python
y=stem_word("this is eated for mooning summered farming")
print(y)
```

**ACTIVATION FUNCTION SIGMOID**

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis("tight")
plt.title("Activation function of sigmoid")
plt.show()
```
