**ACTIVATION FUNCTION SIGMOID**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
def sigmoid(x):
  return (1/(1+np.exp(-x)))
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title("Activation function sigmoid")
#plt.axis("tight")
plt.show()
```

  **ACTIVATION FUNCTION tanh**

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-7,7)
plt.plot(x,tanh(x))
plt.title("ACTIVATION FUNCTION tanh")
plt.show()
```

**ACTIVATION FUNCTION ReLU**

```python
def max(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
  return x1
x=np.linspace(-10,20)
plt.plot(x,max(x))
plt.title("ACTIVATION FUNCTION ReLU")
plt.show()
```
