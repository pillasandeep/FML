```python
import matplotlib.pyplot as plt
x=[1,2,3,3,5]
print(len(x))
y=[1,2,3,4,5]
plt.xlabel("c")
plt.ylabel("a")
plt.title("sample graph")
plt.plot(x,y,marker=".",color="b")
#plt.scatter(x,y)
plt.show()
```

```python
x=[1,3,65,87]
y=[54,65,87,87]
plt.scatter(x,y,color="r")
plt.show()
```

```python
import numpy as py
arr=py.array([1,2,3,5,67,87])
arr1=py.array([1,2,3,5,67,87])
plt.plot(arr,arr1,marker=".",c="r")
plt.show()
```

```python
x=py.arange(0,10,9)
x1=x+2
plt.plot(x,x1,marker="^")
plt.show()
```
